1. What is OCR (Optical Character Recognition)?

OCR (Optical Character Recognition) is a technology that helps computers read text from images, scanned papers, or photos. Instead of typing everything by hand, OCR can automatically convert the text into an editable digital format. It makes working with documents much faster and more convenient.

2. Why is Urdu OCR harder than English OCR?

Urdu OCR is more challenging because Urdu is written in a connected script, and the shape of each letter can change depending on where it appears in a word. Many Urdu letters also look almost the same, with only small dots making the difference. Because of this, along with different fonts and handwriting styles, it is much harder for a computer to recognize Urdu text accurately.

3. What are 2 real-world situations where Urdu OCR would be useful?

Urdu OCR would be very useful for converting old Urdu books, newspapers, and historical records into digital text so they can be searched and preserved. It can also help businesses and government offices quickly extract text from Urdu forms, invoices, or official documents without entering everything manually. This saves time and reduces mistakes.

```python
from PIL import Image, ImageDraw, ImageFont
import arabic_reshaper
from bidi.algorithm import get_display
import os

urdu_texts = [
    "پاکستان زندہ باد",
    "آج کا موسم خوشگوار ہے",
    "تعلیم ہر انسان کا حق ہے",
    "کراچی پاکستان کا سب سے بڑا شہر ہے",
    "محنت کامیابی کی کنجی ہے",
    "علم روشنی ہے",
    "وقت بہت قیمتی ہے",
    "محنت کا پھل میٹھا ہوتا ہے",
    "ہم سب پاکستانی ہیں",
    "صاف ماحول صحت کی ضمانت ہے",
    "درخت لگائیں ماحول بچائیں",
    "کتاب بہترین دوست ہے",
    "ہر انسان برابر ہے",
    "سچ ہمیشہ کامیاب ہوتا ہے",
    "پاکستان ایک خوبصورت ملک ہے",
    "اردو ہماری قومی زبان ہے",
    "پڑھائی کامیابی کی پہلی سیڑھی ہے",
    "ہمیں دوسروں کی مدد کرنی چاہیے",
    "دیانت داری بہترین پالیسی ہے",
    "اتحاد ایمان اور نظم"
]

os.makedirs("data/synthetic", exist_ok=True)

font = ImageFont.truetype(
    "/usr/share/fonts/truetype/noto/NotoNaskhArabic-Regular.ttf",
    32
)

for i, text in enumerate(urdu_texts):
    reshaped = arabic_reshaper.reshape(text)
    bidi_text = get_display(reshaped)

    img = Image.new("RGB", (700, 120), "white")
    draw = ImageDraw.Draw(img)

    bbox = draw.textbbox((0, 0), bidi_text, font=font)
    text_width = bbox[2] - bbox[0]
    text_height = bbox[3] - bbox[1]

    x = 700 - text_width - 20
    y = (120 - text_height) // 2

    draw.text((x, y), bidi_text, fill="black", font=font)

    img.save(f"data/synthetic/urdu_{i+1}.png")

print(" 20 Synthetic Images Generated Successfully!")
```
